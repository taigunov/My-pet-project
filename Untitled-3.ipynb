```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Данные: возраст (X) и надои за сезон (Y)
X = np.array([13, 11, 6, 8, 10, 6, 4, 14, 2]).reshape(-1, 1)
Y = np.array([146.7, 231.9, 269.4, 134.7, 433.4, 173.6, 283.1, 136.5, 35.5])
```

```python
# Построение модели
model = LinearRegression()
model.fit(X, Y)
```

```python
# Прогнозирование
Y_pred = model.predict(X)
```

```python
# График
plt.scatter(X, Y, color='blue', label='Данные')
plt.plot(X, Y_pred, color='red', label='Регрессия')
plt.xlabel('Возраст')
plt.ylabel('Молочная продуктивность за сезон, кг')
plt.title('Влияние возраста на надои лосих')
plt.legend()
plt.show()
```

```python
# Коэффициенты
print(f"Коэффициент a (влияние возраста): {model.coef_[0]}")
print(f"Свободный член b: {model.intercept_}")
```

```python
import statsmodels.api as sm
```

```python
regression = sm.OLS (X,Y_pred)
result = regression.fit()
result.summary()
```
