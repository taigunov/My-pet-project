```python
returns = np.log (data).diff()
returns.head ()
```

```python
returns.dropna(inplace=True)
```

```python
returns.corr
```

```python
sample= returns.sample(60).corr()
sample
```

```python
sample = returns.sample(60)
plt.scatter(x=sample['Milk yield 305, kg - 1 Lac.'], y=sample['Milk yield for the entire lactation, kg - 1 lac.']);
```

```python
reg = np.polyfit(sample['Milk yield 305, kg - 1 Lac.'], sample['Milk yield for the entire lactation, kg - 1 lac.'], deg=1)
reg
```

```python
trend = np.polyval(reg, sample['Milk yield 305, kg - 1 Lac.'])
plt.scatter(sample['Milk yield 305, kg - 1 Lac.'], sample['Milk yield for the entire lactation, kg - 1 lac.'])
plt.plot(sample["Milk yield 305, kg - 1 Lac."], trend, 'r');
```

```python
regression = sm.OLS (sample ['Milk yield for the entire lactation, kg - 1 lac.'], sample ['Milk yield 305, kg - 1 Lac.'])
result = regression.fit()
result.summary()
```
